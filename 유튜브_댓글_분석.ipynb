```python
# ì „ì²˜ë¦¬ ì „ ì›ë³¸ ë³´ì¡´
import copy
copy_data = copy.deepcopy(df)
copy_data.info()
```

```python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)

#ë¶„ì„ì— ì–´ê¸‹ë‚˜ëŠ” ë¶ˆìš©ì–´êµ¬ ì œì™¸ (íŠ¹ìˆ˜ë¬¸ì, ì˜ì„±ì–´)
han = re.compile(r'[ã„±-ã…ã…-ã…£!?~,".\n\r#\ufeff\u200d]')
```

```python
comment_list = []
for i in trange(len(copy_data['comment'])):
  comment_list.append(copy_data.comment.iloc[i])
```

```python
comment_result = []

for i in comment_list:
    tokens = re.sub(emoji_pattern,"",i)
    tokens = re.sub(han,"",tokens)
    comment_result.append(tokens)

comment_result = pd.DataFrame(comment_result, columns=["comment"])
```

```python
import fasttext
model = fasttext.load_model('/content/drive/MyDrive/ML Deep/á„Œá…¡á„‹á…§á†«á„‹á…¥á„á…¥á„…á…µ/lid.176.ftz')

predict = []
for t in comment_result.comment:
  predict.append(model.predict(t,k=1))
```

```python
ty = pd.DataFrame(predict)

comment = []
for num, txt in enumerate(ty[0]):
  #print(txt)
  #if num == 30:
   # break
  
  txt = str(txt)

  if txt == "('__label__ko',)":
    b = re.sub(txt,"ko",txt)
    comment.append(b)
  elif txt == "('__label__en',)":
    b = re.sub(txt,"en",txt)
    comment.append(b)
  elif txt == "('__label__id',)":
    b = re.sub(txt,"id",txt)
    comment.append(b)
  else:
    b = re.sub(txt,"etc",txt)
    comment.append(b)


comment[:10]
comment = pd.DataFrame(comment)
comment
```
